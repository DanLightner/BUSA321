```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

df = pd.read_csv('college91.csv')
df
```

```python
label = "Enrollment"    # this will be your response variable. What are you trying to predict? Plug-and-chug

y = df[label]
X = df.select_dtypes(np.number).assign(const=1)  #drop all categorical features and allow y-intercept to vary
X = X.drop(columns=[label,"Group","Resident"])                       # X is capitalized because it represents a matrix of multiple x-columns

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

pokemon = pd.read_csv('pokemon.csv')
pokemon
```

```python
label = "hp"    # this will be your response variable. What are you trying to predict? Plug-and-chug

y = pokemon[label]
X = pokemon.select_dtypes(np.number).assign(const=1)  #drop all categorical features and allow y-intercept to vary
X = X.drop(columns=[label,"generation", "speed"])                       # X is capitalized because it represents a matrix of multiple x-columns

model = sm.OLS(y, X)
results = model.fit()

print(results.summary())
```
