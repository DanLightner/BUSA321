**T-Tests!** 

Chapter 5 covers analysis of **categorical** with **numerical** variables (or features). For *t-tests*, we are finding a relationship between **one categorical variable** with **one numerical variable** by:

1. **Categorical variable**: one variable with two categories (male/female, smoker/nonsmoker)
2. **Numerical variable**: splitting this set of numbers by the categories above

In other words, we compare the **means** of the numbers between each category. For example:

1. Is there a difference between the mean BMI for men and women?
2. Is there a difference between the mean charges incurred on insurance by smokers and nonsmokers?

**The relationship:** If there's a *significant difference between the two means* based on the category, then the categorical variable affects the numerical variable--the two variables are "related."

**T-test**: As the difference in the mean of each category increases, then so does the t-statistic. As the t-statistic is larger, then the *p-value gets smaller* (based on how much data we have). 

**Meaning (sort of)**: The probability of having a such huge difference in the means between two categories is REALLY REALLY LOW. So, if your dataset happens to show that kind of difference, then you will get a low p-value (sort of)

**ACTUAL Meaning**: If there is TRULY no difference between the categories in real life, then the probability that your dataset DOES show a difference is REALLY LOW, and unlikely. This means that you can be confident (or 95% confident) that your dataset represents the wider population in showing a difference in the categories (phew!)

**Interpretation: p-value < 0.05 means you found something**

```python
from scipy import stats   # import stats package in order to use the t-test in Python
import pandas as pd     # important pandas package in order to create a dataframe of the insurance.csv dataset

Ins = pd.read_csv('http://www.ishelp.info/data/insurance.csv')
bmi = Ins.bmi
charges = Ins.charges

# create some variables for you to use easily
# Each of the following is pulling specific data from the dataset and saving it as a variable

malebmi = bmi[(Ins.sex=='male')]     # remember, you are choosing specific data FROM the BMI feature--only when sex=male
femalebmi = bmi[(Ins.sex=='female')]

smokerBMI = bmi[(Ins.smoker=='yes')]
nonsmokerBMI = bmi[(Ins.smoker=='no')]

smokercharges = charges[(Ins.smoker=='yes')]
nonsmokercharges = charges[(Ins.smoker=='no')]



```

Run some t-tests. The actual code is as follows: stats.ttest_ind(1stcategory,2ndcategory)

Notice that these are two lists of data containing all the values for one category and for the other. Look closely at what your variable codes above are doing.

Which p-values are less than 0.05? Does this make sense? KNOW THY DATA! Use common sense also.

```python
print("Is there a difference in BMI between men and women?\n",stats.ttest_ind(malebmi,femalebmi))
print("Is there a difference in BMI for smokers and non-smokers?\n",stats.ttest_ind(smokerBMI, nonsmokerBMI))
print("Is there a difference in BMI between men and women?\n",stats.ttest_ind(malebmi,femalebmi))
print("Is there a difference in BMI for smokers and non-smokers?\n",stats.ttest_ind(smokerBMI, nonsmokerBMI))
print("Is there a difference in the charges on insurance for smokers and non-smokers?\n",stats.ttest_ind(smokercharges, nonsmokercharges))



print("Is there a difference in the charges on insurance for smokers and non-smokers?\n",stats.ttest_ind(smokercharges, nonsmokercharges))
```

**BMI between men and women**: p-value is 0.08, which means we can only say with 90% confidence (not 95%) that there's a difference in BMI between men and women. This is not that strong, but think about weight and muscle mass tend to be higher for men than women (think about men and women you see on campus, football players, etc.)

**BMI between smokers and non-smokers**: p-value 0.89 -- DEFINITELY no difference there. This makes sense because sometimes smokers have a loss of appetite due to dead taste buds.
