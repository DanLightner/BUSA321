```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
from matplotlib import pyplot as plt

df = pd.read_csv("company.csv")
df = df.drop(columns="Company")
df
```

```python
df.sector.unique()
```

```python
y = df.sector # Label
X = df.drop(columns=['sector']) # Features
X = X.select_dtypes(np.number)
X

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
X_train
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
    
# Train Decision Tree Classifier
clf = clf.fit(X_train,y_train)
    
# Predict the labels for test dataset
y_pred = clf.predict(X_test)
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()
    
# Train Decision Tree Classifier
clf = clf.fit(X_train,y_train)
    
# Predict the labels for test dataset
y_pred = clf.predict(X_test)
```

```python
# View the predicted versus actual in a DataFrame

output_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred,})
output_df
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)

sector_categories = df.sector.unique()
# Display confusion matrix with dynamic labels
cm_display = metrics.ConfusionMatrixDisplay(cm, display_labels=sector_categories)
cm_display.plot(values_format='d')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()

# Print value counts to verify the plot
print(y_test.value_counts())
```

```python

sector_categories = y_test.unique()  # This will get the unique categories from y_test

# One-hot encode both y_test and y_pred
y_test_dummies = pd.get_dummies(y_test)
y_pred_dummies = pd.get_dummies(y_pred)

# Align the columns of y_pred_dummies with y_test_dummies (this ensures both have the same categories)
y_pred_dummies = y_pred_dummies.reindex(columns=y_test_dummies.columns, fill_value=0)

# Generate the classification report with dynamic target names
report_dict = metrics.classification_report(
    y_test_dummies, 
    y_pred_dummies, 
    target_names=sector_categories,  # Use sector categories here
    output_dict=True
)

# Convert the report to a DataFrame for better visualization
df_report = pd.DataFrame(report_dict)
df_report
```

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from io import StringIO
import pydotplus

# Get the unique sector categories dynamically from y_train (or y_test)
sector_categories = np.sort(y_train.unique())  # Get unique labels from training set

# Ensure the number of classes in clf matches the number of classes in sector_categories
if len(sector_categories) != len(clf.classes_):
    print(f"Warning: Number of classes in sector_categories ({len(sector_categories)}) "
          f"does not match the number of classes in clf ({len(clf.classes_)}).")
    print(f"Using the classes from the model instead: {clf.classes_}")
    sector_categories = clf.classes_  # Use classes from the model if there's a mismatch

# Export decision tree to DOT format
dot_data = StringIO()
export_graphviz(clf,  # This is the trained model
                out_file=dot_data,
                filled=True,
                rounded=True,
                special_characters=True,
                feature_names=X.columns,
                class_names=sector_categories)  # Use dynamic class names

# Generate graph from DOT data
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

# Create and display the decision tree image
graph.write_png("decision_tree.png")  # Save the tree image as a PNG
```
