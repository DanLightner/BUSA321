```python
import matplotlib.pyplot as plt
from scipy import stats

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.scatter(df['tempmax'], df['humidity'], alpha=0.7)
plt.xlabel('Max Temperature (°F)')
plt.ylabel('Humidity (%)')

m1, b1, r1, p1, err1 = stats.linregress(df['tempmax'], df['humidity'])
plt.plot(df['tempmax'], m1 * df['tempmax'] + b1, 'r--')
plt.title(f'Max Temp vs Humidity')

plt.subplot(1, 3, 2)
plt.scatter(df['tempmax'], df['Health_Risk_Score'], alpha=0.7)
plt.xlabel('Max Temperature (°F)')
plt.ylabel('Health Risk Score')
m2, b2, r2, p2, err2 = stats.linregress(df['tempmax'], df['Health_Risk_Score'])
plt.plot(df['tempmax'], m2 * df['tempmax'] + b2, 'r--')
plt.title(f'Max Temp vs Health Risk Score')

plt.subplot(1, 3, 3)
plt.scatter(df['humidity'], df['Health_Risk_Score'], alpha=0.7)
plt.xlabel('Humidity (%)')
plt.ylabel('Health Risk Score')
m3, b3, r3, p3, err3 = stats.linregress(df['humidity'], df['Health_Risk_Score'])
plt.plot(df['humidity'], m3 * df['humidity'] + b3, 'r--')
plt.title(f'Humidity vs Health Risk Score')

plt.tight_layout()

plt.show()
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
from statsmodels.stats.diagnostic import het_white
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols(formula='tempmax~humidity', data=df).fit()
model.params


#TEMP AND HUMIDITY
```

```python
white_test = het_white(model.resid,  model.model.exog)
bp_test = het_breuschpagan(model.resid, model.model.exog)
output_df = pd.DataFrame(columns=['LM stat', 'LM p-value', 'F-stat', 'F p-value'])
output_df.loc['White'] = white_test
output_df.loc['Breusch-Pagan'] = bp_test
output_df.round(5)

#PAGAN/WHITE TEST
```

```python
from scipy.stats import pearsonr
corr_temp_humidity, p_temp_humidity = pearsonr(df['tempmax'], df['humidity'])
corr_temp_health, p_temp_health = pearsonr(df['tempmax'], df['Health_Risk_Score'])
corr_humidity_health, p_humidity_health = pearsonr(df['humidity'], df['Health_Risk_Score'])

print(f'Correlation between tempmax and humidity: {corr_temp_humidity}, | p-value: {p_temp_humidity}')
print(f'Correlation between tempmax and Health Risk Score: {corr_temp_health}, | p-value: {p_temp_health}')
print(f'Correlation between humidity and Health Risk Score: {corr_humidity_health}, | p-value: {p_humidity_health}')
```

**Temperature and Humidity:**

The p-value is very low, meaning we can confidently say that the two are statistically significant. This suggests that as temperature increases, humidity tends to decrease. In terms of assumptions, linearity is present, as the scatter plot shows a downward trend. However, equal variance did not pass the assumption due to the Pagan and White test p-values being so low. **Temperature and Health Risk Score:**
The correlation between temperature (tempmax) and Health Risk Score shows a slightly positive correlation of about 0.35, with a very low p-value, indicating a statistically significant relationship. This suggests that higher temperatures are associated with higher health risk scores; however, it doesn't tell the full story. The assumptions for this correlation appear to be mostly met, as there seems to be a linear relationship but there tends to be a fair amount of outliers. While the data suggests a trend, the potential there is a lot up for debate. The White test indicates significant heteroscedasticity with a very low p-value, suggesting that the spread of the residuals is not consistent throughout the dataset. The Breusch-Pagan test exhibits a high p-value, which indicates that the linear relationship shows equal variance in the residuals, suggesting that in a linear sense, there is a equal variance. This creates confusion because, while there appears to be a linear relationship, the assumption of homoscedasticity—where the residuals have a constant variance—does not hold true.

**Humidity and Health Risk Score:**
For the relationship between humidity and Health Risk Score, the correlation coefficient is approximately 0.19, with a p-value that is also very low. This indicates a weak positive correlation, suggesting that as humidity increases, health risk scores tend to rise slightly. However, this relationship is significantly less strong than the others, and while the linearity assumption is likely met, it’s important to note that the weak correlation indicates that humidity may not be a strong predictor of health risks. Additionally, the assumption of homoscedasticity should be considered, as the variance of the residuals might not be constant across different levels of humidity.
