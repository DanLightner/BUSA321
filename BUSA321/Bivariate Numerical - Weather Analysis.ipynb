```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.scatter(df['tempmax'], df['humidity'])
plt.xlabel('Max Temperature (°F)')
plt.ylabel('Humidity (%)')
plt.title('Max Temperature vs Humidity')

plt.subplot(1, 3, 2)
plt.scatter(df['tempmax'], df['Health_Risk_Score'])
plt.xlabel('Max Temperature (°F)')
plt.ylabel('Health Risk Score')
plt.title('Max Temperature vs Health Risk Score')

plt.subplot(1, 3, 3)
plt.scatter(df['humidity'], df['Health_Risk_Score'])
plt.xlabel('Humidity (%)')
plt.ylabel('Health Risk Score')
plt.title('Humidity vs Health Risk Score')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import pearsonr
corr_temp_humidity, p_temp_humidity = pearsonr(df['tempmax'], df['humidity'])
corr_temp_health, p_temp_health = pearsonr(df['tempmax'], df['Health_Risk_Score'])
corr_humidity_health, p_humidity_health = pearsonr(df['humidity'], df['Health_Risk_Score'])

print(f'Pearson correlation between tempmax and humidity: {corr_temp_humidity}, p-value: {p_temp_humidity}')
print(f'Pearson correlation between tempmax and Health_Risk_Score: {corr_temp_health}, p-value: {p_temp_health}')
print(f'Pearson correlation between humidity and Health_Risk_Score: {corr_humidity_health}, p-value: {p_humidity_health}')
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.diagnostic import het_breuschpagan, het_white

model = ols(formula='Health_Risk_Score ~ tempmax + humidity', data=df).fit()

print(model.params)
white_test = het_white(model.resid, model.model.exog)

bp_test = het_breuschpagan(model.resid, model.model.exog)

output_df = pd.DataFrame(columns=['LM stat', 'LM p-value', 'F-stat', 'F p-value'])
output_df.loc['White'] = white_test
output_df.loc['Breusch-Pagan'] = bp_test
output_df = output_df.round(6)
print(output_df)
```
