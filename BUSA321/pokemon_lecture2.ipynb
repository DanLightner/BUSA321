```python
import seaborn as sns
import pandas as pd
df = pd.read_csv("pokemon.csv")
df
```

```python
sns.jointplot(data=df, x='defense', y='attack')
```

```python
sns.jointplot(data=df, x='weight_kg', y='defense')
```

```python
sns.jointplot(data=df, x='attack', y='defense', kind="hex", color="k");
```

```python
import matplotlib.pyplot as plt
sns.set_style('white')
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(x='attack', y='defense', data=df, ax=ax)
sns.rugplot(x='attack', data=df, color="g", ax=ax)
sns.rugplot(y='defense', data=df, ax=ax);
```

```python
sns.set(rc={'axes.facecolor':'black', 'figure.facecolor':'white', 'axes.grid':False})
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0.05, light=1, reverse=True)
sns.kdeplot(x='attack', y='defense', data=df, cmap=cmap, n_levels=60, shade=True);
```

```python
sns.set_style('dark')
g = sns.jointplot(x="attack", y="defense", data=df, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$X$", "$Y$");
```

```python
df.dropna(inplace=True) # remove null values first
sns.pairplot(df);
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=60);
```
