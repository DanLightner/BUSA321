Import the imdb.csv data file into a Pandas DataFrame and name it **df**

```python

import pandas as pd

df = pd.read_csv('imdb.csv')
df
```

The code in the code cell below generates a unique list of genres based on each of those that appear in the Genre column. 

```python
genres = []

for row in df.Genre:            # for each row in the Genre column
  for genre in row.split(','):  # for each genre in the row
    if genre not in genres:     # if that genre is not already in the unique genre list
      genres.append(genre)      # add it to the list
      
genres
```

**NEXT**: Let's analyze the effect of genre on Rating. Create 5 lists of Ratings by **filling in the following code**--one list for each of the following genres: Action, Drama, Comedy, Horror, Romance

Compare these lists of ratings in an ANOVA to see if there are any significant differences. Print out the F and p-value from this ANOVA.

```python
from scipy import stats
import statistics as stat

# Extract lists of ratings for each genre
action = df[df['Action'] == True]['Rating']
drama = df[df['Drama'] == True]['Rating']
comedy = df[df['Comedy'] == True]['Rating']
horror = df[df['Horror'] == True]['Rating']
romance = df[df['Romance'] == True]['Rating']

# Perform ANOVA
F, p = stats.f_oneway(action, drama, comedy, horror, romance)

# Print F and p-values
print("F-value:", F)
print("p-value:", p)
```

**Next**: Print the means of each of the genre ratings.

The results from the ANOVA test show that the F-value is 20.29, and the p-value is extremely small, about 3.63e-16. This suggests there are significant statistical differences in the average ratings between the different genres: Action, Drama, Comedy, Horror, and Romance. Since the p-value is so small, it indicates that it’s highly unlikely these differences happened by chance. In other words, the genre of a movie seems to have a noticeable impact on its average rating. We would need to explore further to see exactly which genres are driving these differences, but overall, it's clear that not all genres have the same average ratings.

```python
# Calculate and print the mean ratings for each genre
mean_action = df[df['Action'] == True]['Rating'].mean()
mean_drama = df[df['Drama'] == True]['Rating'].mean()
mean_comedy = df[df['Comedy'] == True]['Rating'].mean()
mean_horror = df[df['Horror'] == True]['Rating'].mean()
mean_romance = df[df['Romance'] == True]['Rating'].mean()

print("Mean Ratings:")
print(f"Action: {mean_action:.2f}")
print(f"Drama: {mean_drama:.2f}")
print(f"Comedy: {mean_comedy:.2f}")
print(f"Horror: {mean_horror:.2f}")
print(f"Romance: {mean_romance:.2f}")
```
