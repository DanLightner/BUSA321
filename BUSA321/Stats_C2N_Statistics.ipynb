Import the imdb.csv data file into a Pandas DataFrame and name it **df**

```python
import pandas as pd
df = pd.read_csv('imdb.csv')
df
```

The code in the code cell below generates a unique list of genres based on each of those that appear in the Genre column. 

```python
genres = []

for row in df.Genre:            # for each row in the Genre column
  for genre in row.split(','):  # for each genre in the row
    if genre not in genres:     # if that genre is not already in the unique genre list
      genres.append(genre)      # add it to the list
      
genres
```

**NEXT**: Let's analyze the effect of genre on Rating. Create 5 lists of Ratings by **filling in the following code**--one list for each of the following genres: Action, Drama, Comedy, Horror, Romance

Compare these lists of ratings in an ANOVA to see if there are any significant differences. Print out the F and p-value from this ANOVA.

```python
from scipy import stats
import statistics as stat
action = df[df['Action'] == True]['Rating']
drama = df[df['Drama'] == True]['Rating']
comedy = df[df['Comedy'] == True]['Rating']
horror = df[df['Horror'] == True]['Rating']
romance = df[df['Romance'] == True]['Rating']
F, p = stats.f_oneway(action, drama, comedy, horror, romance)


print("F-value:", F)
print("p-value:", p)
```

Since the p-value is so small, it indicates that itâ€™s highly unlikely these differences happened by chance. **NEXT**: Let's see if the critics agree with the movie-goers. **Perform the same analyses using Metascore** (a score generated by critics) as the label. Compare the same five genres and print out both their mean Metascores. **Comment on your results**

These results highlight a big contrast in how critics see the quality of films based on genre. **NEXT**: Let's run one more analyses on the Year of the movie. Even though Year is numeric and has an order, we do not have a theory (i.e. a logical reason) to suspect that movies get better or worse over time. Therefore, we should **analyze Year as a nominal/object** data type using an ANOVA. 

Therefore, convert the Year column to an object and then compare the Rating of the years 2012-2016 by **completing the following code** to print out both the mean Rating for each of those years and the F and p-value of an ANOVA comparing them. **Comment on your analysis**

```python

label = 'Rating'

# Extract ratings for each year
y2012 = df[df['Year'] == '2012'][label]
y2013 = df[df['Year'] == '2013'][label]
y2014 = df[df['Year'] == '2014'][label]
y2015 = df[df['Year'] == '2015'][label]
y2016 = df[df['Year'] == '2016'][label]

F_year, p_year = stats.f_oneway(y2012, y2013, y2014, y2015, y2016)

mean_y2012 = y2012.mean()
mean_y2013 = y2013.mean()
mean_y2014 = y2014.mean()
mean_y2015 = y2015.mean()
mean_y2016 = y2016.mean()

print("Mean Ratings:")
print("2012:", mean_y2012)
print("2013:", mean_y2013)
print("2014:", mean_y2014)
print("2015:", mean_y2015)
print("2016:", mean_y2016)

print("\nANOVA Results:")
print("F-value:", F_year)
print("p-value:", p_year)
```

The ANOVA test results reveal an F-value of 2.11 and a p-value of 0.079. Since the p-value is slightly above the conventional threshold of 0.05, we cannot conclude that there are statistically significant differences in average movie ratings across the years 2012 to 2016. When examining the mean ratings, 2012 stands out with the highest average at 6.93, while 2016 has the lowest at 6.64. Although these differences suggest a slight decline in ratings over the years, the lack of statistical significance indicates that these variations may not be substantial enough to draw strong conclusions. Overall, while the data shows some variation in ratings year by year, it doesn't provide robust evidence that ratings significantly differ based on the year alone. Perhaps there might be other factors to suggest this idea.
